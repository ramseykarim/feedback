```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
mega_gs = fig.add_gridspec(right=0.9)
mega_ax = mega_gs.subplots()
gs = mega_gs[0,0].subgridspec(2, 2)
axes = gs.subplots()
for i in range(2):
    for j in range(2):
        x = np.arange(10)
        axes[i,j].plot(x*(1 if i else -1), x*(1 if j else -1))
#         axes[i,j].set_axis_off()
cbar_ax = mega_ax.inset_axes([1.05, 0, 0.05, 1])
cbar_ax.imshow(np.arange(500).reshape(100, 5))
# cbar_ax.set_axis_off()
mega_ax.set_axis_off() # this is the solution!
```
