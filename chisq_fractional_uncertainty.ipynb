# Chi squared ellipse and fractional uncertainties

January 3, 2024

I want to check if the chi squared surface between two independent variables can be dominated by one measurement with a small fractional error, and how small that fractional error needs to be.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid_params = (-5, 5, 20)
```

```python
xx, yy = np.meshgrid(np.linspace(*grid_params), np.linspace(*grid_params))
xerr = 3
yerr = 3

chisq = ((xx-0)/xerr)**2 + ((yy-0)/yerr)**2 + 1

plt.imshow(np.log10(chisq), extent=[-5, 5, -5, 5])
plt.colorbar(label='log10 chi sq')
plt.contour(chisq, levels=[np.min(chisq)*2], colors='w', extent=[-5, 5, -5, 5])
```

```python
xx, yy = np.meshgrid(np.linspace(*grid_params), np.linspace(*grid_params))
xerr = 0.3
yerr = 3

chisq = ((xx-0)/xerr)**2 + ((yy-0)/yerr)**2 + 1

plt.imshow(np.log10(chisq), extent=[-5, 5, -5, 5])
plt.colorbar(label='log10 chi sq')
plt.contour(chisq, levels=[np.min(chisq)*2], colors='w', extent=[-5, 5, -5, 5])
```
